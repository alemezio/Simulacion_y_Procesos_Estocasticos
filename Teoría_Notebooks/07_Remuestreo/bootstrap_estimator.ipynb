```python
import random
import numpy as np 
import matplotlib.pyplot as plt
```

Población: 1000 elementos tomados de una distribución uniforme entre 0 y 50

Muestra: 100 elementos tomados de esa población

```python
random.seed(7)

PopData = list()

for i in range(1000):
    DataElem = 50 * random.random()
    PopData.append(DataElem)
```

```python
PopSample = random.choices(PopData, k=100)

PopSampleMean = list()
for i in range(10000):  
    # SampleI = random.choices(PopData, k=100)
    SampleI = random.choices(PopSample, k=100)
    PopSampleMean.append(np.mean(SampleI))
```

```python
plt.hist(PopSampleMean,bins=20)
plt.show()

MeanPopSample = np.mean(PopSample)
print("The mean of the original sample is ",MeanPopSample)

MeanPopData = np.mean(PopData)
print("The mean of the population is ",MeanPopData)

MeanPopSampleMean = np.mean(PopSampleMean)
print("The mean of the Bootstrap estimator is ",MeanPopSampleMean)
```

```python
print("The standard deviation of the Bootstrap estimator is ",np.std(PopSampleMean))

print("The standard deviation of the original sample is ",np.std(PopSample))
```
