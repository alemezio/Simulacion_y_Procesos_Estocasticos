```python
from src.chi_square_test import chi_square_uniform_test
from src.ks_test import kolmogorov_smirnov_uniform_test
import numpy as np
```

```python
np.random.seed(42)

# Test with uniform distributed data
uniform_data = np.random.uniform(0, 1, 1000)
result_uniform_chi = chi_square_uniform_test(uniform_data)
print("Test with True Uniform Data:")
print(result_uniform_chi)
```

Chequeemos calculando la distribución Chi-cuadrado con 9=10-1 grados de libertad, para una significancia alpha=0.05, y veamos que es un valor mayor 9.4:

```python
from scipy.stats import chi2

# Critical value for alpha = 0.05
cv = chi2.ppf(0.95, df=9)
print(f"Critical value (α=0.05): {cv:.4f}")
```

También, chequeamos que la distribución Chi-cuadrado con 9 grados de libertad evaluada en x=9.4 tiene un p-value mayor que alpha:

```python
x = 9.4
k = 9

pdf  = chi2.pdf(x, df=k)
cdf  = chi2.cdf(x, df=k)
pval = chi2.sf(x, df=k)   # survival function = 1 - CDF

print(f"PDF:     {pdf:.5f}")
print(f"CDF:     {cdf:.4f}")
print(f"p-value: {pval:.4f}")
```

Test Kolmogorov-Smirnov:

Ahora los dos tests pero con una distribución no uniforme (usamos la normal), por lo que ambos deberían rechazar la hipótesis nula:

```python
# Test with normal distributed data
normal_data = np.random.normal(0.5, 0.1, 1000)
# Clip normal data between 0 and 1 so it fits the range, but clearly not uniformly distributed
normal_data = np.clip(normal_data, 0, 1) 
result_normal_chi = chi_square_uniform_test(normal_data)
print("Test with Non-Uniform (Normal) Data:")
print(result_normal_chi)
```

```python
result_normal_ks = kolmogorov_smirnov_uniform_test(normal_data)
print("Test with Non-Uniform (Normal) Data:")
print(result_normal_ks)
```
