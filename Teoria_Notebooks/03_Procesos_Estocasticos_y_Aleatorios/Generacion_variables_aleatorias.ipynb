# Generación de números aleatorios con diferentes distribuciones de probabilidad

## Discretas

```python
from random import random
```

```python
# x: vector de valores posibles de X
# p: vector de probabilidades
def discretaX(p, x):
    U = random()
    i, F = 0, p[0]
    while U >= F:
        i += 1
        F += p[i]
    return x[i]
```

```python
p=[0.1, 0.2, 0.3, 0.4]
x=[1, 2, 3, 4]
print(discretaX(p, x))
```

```python
# Generación de una variable aleatoria discreta X con distribución dada por:
# P(X=1) = 0.20, P(X=2) = 0.15, P(X=3) = 0.25, P(X=4) = 0.40
def generaX1():
    U = random()
    if U < 0.20:   return 1
    elif U < 0.35: return 2
    elif U < 0.60: return 3
    else:          return 4
```

```python
print(generaX1())
print(generaX1())
print(generaX1())
```

### Generación de distribuciones particulares

Uniforme:

```python
def udiscreta(n):
    U = random()
    x = 1
    F = 1 / n
    while U >= F:
        F += 1 / n
        x += 1
    return x
```

Geométrica:

```python
def geometrica(p):
    U = random()
    return int(log(1-U)/log(1-p))+1
```

Bernoulli:

```python
def Bernoulli(p):
    U=random()
    if U < p: return 1
    else:     return 0
```

Poisson:

```python
def Poisson(lamda):
    U = random()
    i = 0
    p = exp(-lamda)
    F = p
    while U >= F:
        i += 1
        p *= lamda / i
        F = F + p
    return i
```

Binomial:

```python
def Binomial(n,p):
    c = p / (1 - p)
    prob = (1 - p) ** n
    F = prob
    i = 0
    U = random()
    while U >= F:
        prob *= c * (n-i) / (i+1)
        F += prob
        i += 1
    return i
```
