Veamos una simple implementación del Generador de Numeros Aleatorios de von Neumann

```python
def vonNeumann(x0,n):
    u = x0
    print('semilla: ',u)
    for i in range(n): 
        u= (u**2 // 100) % 10000
        yield u
```

```python
for i in vonNeumann(1234,10):
    print(i)
```

```python
for i in vonNeumann(2100,8):
    print(i)
```

Generador congruencial lineal

```python
def lcg(m,a,c,x0,n):
    x = x0
    print('semilla: ',x)
    for i in range(n):
        x = (a*x + c) % m
        yield x
```

```python
def lcg_norm(m,a,c,x0,n):
    x = x0
    print('semilla: ',x)
    for i in range(n):
        x = (a*x + c) % m
        u = x/m
        yield u
```

```python
for i in lcg(16,5,3,7,10):
    print(i)
```

```python
for i in lcg_norm(16,5,3,7,10):
    print(i)
```

```python
for i in lcg(13,12,0,1,10):
    print(i)
```

```python
for i in lcg_norm(2**31-1,75,0,0.1,30):
    print(i)
```

Generador de Fibonacci rezagado

```python
import numpy as np
x0=1
x1=1
m=2**32

for i in range (1,21):
    # x= np.mod((x0+x1), m)
    x = (x0+x1) % m
    x0=x1
    x1=x
    print(x)
```

```python
import random
```

```python
for i in range(20):
    print('%05.4f' % random.random(), end=' ')
print()
```

```python
random.seed(1)

for i in range(20):
    print('%05.4f' % random.random(), end=' ')
print()
```

```python
for i in range(20):
    print('%6.4f' %random.uniform(1, 100), end=' ')
print()
```

```python
for i in range(20):
    print(random.randint(-100, 100), end=' ')
print()
```

```python
for i in range(20):
    print(random.randrange(0, 100,5), end=' ')
print()
```
