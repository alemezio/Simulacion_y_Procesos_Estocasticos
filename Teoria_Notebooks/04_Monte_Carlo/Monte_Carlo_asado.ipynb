```python
import numpy as np

sims=1000000

A = np.random.uniform(1,5,sims)
B = np.random.uniform(2,6,sims)

duration = A + B
```

```python
import matplotlib.pyplot as plt

plt.hist(duration, density=True)
plt.axvline(x=9, color='r')
plt.show()
print(f"Duración promedio: {duration.mean():.2f}")
print(f"Probabilidad llegar tarde: {(duration > 9).sum()/sims:.4f}")
```

12.5% de probabilidad de llegar tarde
