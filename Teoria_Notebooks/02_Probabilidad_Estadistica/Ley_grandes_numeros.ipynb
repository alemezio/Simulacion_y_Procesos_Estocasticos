```python
import random

def simulate_coin_flips(num_flips):
    heads = sum(random.choice([0, 1]) for _ in range(num_flips))
    frequency = (heads / num_flips) * 100
    return heads, frequency

def print_table(results):
    col1 = "Number of coin flips"
    col2 = "Number of heads"
    col3 = "Head output frequency"

    w1, w2, w3 = len(col1) + 2, len(col2) + 2, len(col3) + 2

    def row_sep():
        return f"+{'-' * w1}+{'-' * w2}+{'-' * w3}+"

    def row(c1, c2, c3):
        return f"| {str(c1):<{w1-2}} | {str(c2):<{w2-2}} | {str(c3):<{w3-2}} |"

    print(row_sep())
    print(row(col1, col2, col3))
    print(row_sep())
    for flips, heads, freq in results:
        print(row(f"{flips:,}", heads, f"{freq:.1f}%"))
    print(row_sep())

flip_counts = [10, 100, 1_000, 10_000, 100_000, 1_000_000]

results = []
for n in flip_counts:
    heads, freq = simulate_coin_flips(n)
    results.append((n, heads, freq))

print("\nCoin Flip Simulation Results\n")
print_table(results)
```
